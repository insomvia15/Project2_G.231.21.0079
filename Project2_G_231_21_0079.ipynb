```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Membaca dataset dari file CSV
data = pd.read_csv('project2a.csv')

# Ekstraksi data dari dataset
periode_tahun = data.columns.tolist()
periode_bulan = np.arange(1, len(data) + 1)
curah_hujan = data.values

# Membuat subplot
fig, ax = plt.subplots(2, 1, gridspec_kw={'height_ratios': [3, 1]}, figsize=(12, 6))

# Pilihan warna untuk batang
warna = ['blue', 'red', 'lime']

# Grafik batang untuk periode tahun dan bulan
for i in range(len(periode_tahun)):
    ax[0].bar(periode_bulan + i * 0.2, curah_hujan[:, i], width=0.2, label=periode_tahun[i], color=warna[i])

ax[0].legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=len(periode_tahun), frameon=False)
ax[0].set(xticks=periode_bulan + 0.2, xticklabels=periode_bulan)
ax[0].grid(axis='y', color='gray', linestyle='-', linewidth=0.5)
ax[0].set_ylabel('Curah Hujan (mm)', fontsize=16, fontweight='bold')

# Garis pembatas
ax[0].axhline(y=0, color='black', linewidth=0.5)

# Label bulan
ax[1].text(0.5, 0.5, 'Bulan', fontsize=16, fontweight='bold', horizontalalignment='center', verticalalignment='center')
ax[1].axis('off')

# Label Curah Hujan (mm)
curah_hujan_labels = np.arange(700, -1, -100)
ax[0].set_yticks(curah_hujan_labels)
ax[0].set_yticklabels(curah_hujan_labels)

# Menampilkan grafik
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Membaca dataset dari file CSV
df = pd.read_csv('project2b.csv')

# Mengatur format waktu
date_range = pd.date_range(start='2013-07-14 00:00', end='2013-07-15 00:00', freq='3H')
time_labels = date_range.strftime('%d-%b %H:%M UTC')

# Mengatur label pada sumbu X
x_ticks = np.linspace(0, len(df)-1, len(date_range))
x_labels = ['Asimilasi', 'Observasi', 'Non Asimilasi']

# Mengatur label pada sumbu Y
y_ticks = range(27, 19, -1)
y_labels = [str(y) + '°C' for y in y_ticks]

# Membuat plot menggunakan matplotlib
fig, ax = plt.subplots()
fig.set_size_inches(12, 6)

# Menampilkan garis untuk setiap jenis data
ax.plot(x_ticks, df['Asimilasi'], marker='o', color='blue', label='Asimilasi')
ax.plot(x_ticks, df['Observasi'], marker='o', color='gold', label='Observasi')
ax.plot(x_ticks, df['Non Asimilasi'], marker='o', color='red', label='Non Asimilasi')

# Menampilkan label pada sumbu X
ax.set_xticks(x_ticks)
ax.set_xticklabels(time_labels, rotation=45, ha='right')

# Menampilkan label pada sumbu Y
ax.set_yticks(y_ticks)
ax.set_yticklabels(y_labels)

# Menambahkan grid pada sumbu Y
ax.yaxis.grid(True)

# Menampilkan judul dan label
plt.title('Suhu Permukaan')
plt.xlabel('Waktu')
plt.ylabel('Celcius')

# Menampilkan legenda
ax.legend(bbox_to_anchor=(1.18, 0.5), loc='center right', frameon=False)

# Menampilkan grafik
plt.tight_layout()
plt.show()
```
